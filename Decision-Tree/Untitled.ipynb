```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing

features = pd.read_csv('test2.csv')
le = preprocessing.LabelEncoder()

# preprocessing

features['employee'] = features['employee'].fillna('N')
features['gender'] = features['gender'].fillna('H')
features['age'] = features['age'].fillna(features['age'].mean())
features['since_first_date'] = features['since_first_date'].fillna(6)
features['cust_type'] = features['cust_type'].fillna(1)
features['res_index'] = features['res_index'].fillna('S')
features['for_idx'] = features['for_idx'].fillna('N')
features = features.drop('sp_index',axis = 1)
features['channel'] = features['channel'].fillna('KFA')
features['res_code'] = features['res_code'].fillna(features['res_code'].mean())
features['active'] = features['active'].fillna(features['active'].mean())
features['income'] = features['income'].fillna(features['income'].mean())
features['cate'] = features['cate'].fillna('01 - TOP')
features = features.fillna(0)

# Convert categorical variable to numeric
for i in ['employee','gender','res_index','res_code','for_idx','channel','cate']:
    features[i] = le.fit_transform(features[i])

print('The shape of our features is : ',features.shape)

labels = np.array(features['fund'])
features = features.drop('fund',axis = 1)
features = np.array(features)

train_features, test_features, train_labels, test_labels = train_test_split(features,labels,test_size = 0.25,random_state = 42)

print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
print(test_labels)

rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
rf.fit(train_features, train_labels)

predictions = rf.predict(test_features)



errors = abs(predictions - test_labels)
mape = 100 * errors

accuracy = 100 - np.mean(mape)
print('Accuracy: ', round (accuracy, 2), '%.')
```
